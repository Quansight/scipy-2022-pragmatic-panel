```python
import param
import numpy as np
import holoviews as hv
import panel as pn

pn.extension(template='fast')

pn.state.template.title = 'URL Syncing'
```

```python
# pn.state.session_args are the query parameters passed to the Tornado server when the initial request

# pn.state.location.search is the live state

# pn.state.location.sync is an easy way to keep a parameter synced with a URL argument
```

```python
slider = pn.widgets.FloatSlider(start=0, end=10).servable()

pn.state.location.sync(slider, {'value': 'my_custom_argument'})
```

```python
class App(param.Parameterized):
    
    frequency = param.Number(bounds=(1, 10), default=5)
    
    amplitude = param.Number(bounds=(1, 10), default=1)
    
    def plot(self):
        xs = np.linspace(0, np.pi)
        ys = np.sin(xs*self.frequency)*self.amplitude
        return hv.Curve((xs, ys))

app = App()
    
pn.state.location.sync(app)

pn.Param(app.param).servable(area='sidebar')
pn.panel(app.plot).servable(area='main')
```
